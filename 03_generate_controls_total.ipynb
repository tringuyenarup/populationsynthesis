```python
import pandas as pd
import geopandas as gpd
```

```python
area_lev = 'SA2'
# area_lev = 'SA3'
```

```python
file_name_tot_hh = f'2016Census_G30_VIC_{area_lev}.csv' #hh, vel
file_name_tot_pp = f'2016Census_G01_VIC_{area_lev}.csv' #genders, pp, age gr
file_name_tot_work = f'2016Census_G43B_VIC_{area_lev}.csv' #work
file_name_tot_hh2 = f'2016Census_G31_VIC_{area_lev}.csv' #hh size
```

```python
df_hh = pd.read_csv('inputs/melbourne/census/2016_census/'+area_lev+'/VIC/'+file_name_tot_hh)
df_hh2 = pd.read_csv('inputs/melbourne/census/2016_census/'+area_lev+'/VIC/'+file_name_tot_hh2)
df_pp = pd.read_csv('inputs/melbourne/census/2016_census/'+area_lev+'/VIC/'+file_name_tot_pp)
df_work = pd.read_csv('inputs/melbourne/census/2016_census/'+area_lev+'/VIC/'+file_name_tot_work)


id_name = df_hh.columns[0]
```

```python
df_hh_select = df_hh[[
        id_name, 
        'Num_MVs_per_dweling_0_MVs', 
        'Num_MVs_per_dweling_1_MVs', 
        'Num_MVs_per_dweling_2_MVs', 
        'Num_MVs_per_dweling_3_MVs', 
        'Num_MVs_per_dweling_4mo_MVs', 
        'Num_MVs_NS', 
        'Total_dwelings'
        ]]
df_hh_select2 = df_hh2[[
        id_name, 
        'Num_Psns_UR_1_Total',
        'Num_Psns_UR_2_Total',
        'Num_Psns_UR_3_Total',
        'Num_Psns_UR_4_Total',
        'Num_Psns_UR_5_Total',
        'Num_Psns_UR_6mo_Total'
        ]]
df_pp_select = df_pp[[
        id_name, 
        'Tot_P_M', 
        'Tot_P_F', 
        'Tot_P_P', 
        'Age_0_4_yr_P', 
        'Age_5_14_yr_P', 
        'Age_15_19_yr_P', 
        'Age_20_24_yr_P', 
        'Age_25_34_yr_P', 
        'Age_35_44_yr_P', 
        'Age_45_54_yr_P', 
        'Age_55_64_yr_P', 
        'Age_65_74_yr_P', 
        'Age_75_84_yr_P', 
        'Age_85ov_P'
        ]]
df_work_select = df_work[[
        id_name,
        'P_Emp_FullT_Tot',
        'P_Emp_PartT_Tot',
        'P_Emp_awy_f_wrk_Tot',
        'P_Hours_wkd_NS_Tot',
        'P_Tot_Unemp_Tot'
        ]]
# log.info("Combine to produce the final dataset")
df_final = df_hh_select2.merge(df_hh_select.merge(df_pp_select.merge(df_work_select, how='inner', on=id_name), how='inner', on=id_name), how='inner', on=id_name)
df_final["STATE_CODE_2016"] = 2
df_final["P_Emp_Guess_Casual_Tot"] = df_final[["P_Emp_awy_f_wrk_Tot", "P_Hours_wkd_NS_Tot"]].sum(axis=1)
```

```python
df_final["P_Tot_Unemp_Tot"] = df_final['Tot_P_P'] - df_final[["P_Emp_FullT_Tot", "P_Emp_PartT_Tot", \
                                                    "P_Emp_Guess_Casual_Tot"]].sum(axis=1)
```

# Merging empty SA4 cells

```python
new_map_SA4 = {
    220: [203, 217],
    221: [213, 201, 215],
    222: [210, 202],
    223: [209, 204, 216],
    224: [212, 205]
}
aff_sa4 = []
for k in new_map_SA4: aff_sa4.extend(new_map_SA4[k])
```

```python
def add_new_SA4_col(df, name_old_col, map_di, name_new_col="newSA4"):
    arr = []
    for old_z in df[name_old_col]:
        if old_z in map_di: arr.append(map_di[old_z])
        else: log.warning(f"this {old_z} is not exist in mapping dict, probably type convert issue")
    df[name_new_col] = arr
    return df
```

```python
household_seed = pd.read_csv('outputs/seed_households.csv')
person_seed = pd.read_csv('outputs/seed_persons.csv')
controls_total_sa4 = pd.read_csv('outputs/controls_total_SA4.csv')
geo_cross_walk = pd.read_csv('outputs/geo_cross_walk.csv')
```

```python
mapping_dict = {}
for z in controls_total_sa4['SA4_CODE_2016']:
    if z not in aff_sa4:
        mapping_dict[z] = z
    else:
        for k in new_map_SA4:
            if z in new_map_SA4[k]:
                mapping_dict[z] = k
```

```python
household_seed = add_new_SA4_col(household_seed, "SA4", mapping_dict)
household_seed = household_seed.drop(["SA4"], axis=1)

person_seed = add_new_SA4_col(person_seed, "SA4", mapping_dict)
person_seed = person_seed.drop(["SA4"], axis=1)
```

```python
geo_cross_walk = add_new_SA4_col(geo_cross_walk, "SA4_CODE_2016", mapping_dict)
geo_cross_walk = geo_cross_walk.drop(["SA4_CODE_2016"], axis=1)
geo_cross_walk = geo_cross_walk.drop(geo_cross_walk[geo_cross_walk['newSA4'] == 299].index)
geo_cross_walk = geo_cross_walk.drop(geo_cross_walk[geo_cross_walk['newSA4'] == 297].index)
```

```python
controls_total_sa4 = add_new_SA4_col(controls_total_sa4, "SA4_CODE_2016", mapping_dict)
controls_total_sa4 = controls_total_sa4.drop(["SA4_CODE_2016"], axis=1)
controls_total_sa4 = controls_total_sa4.groupby(['newSA4']).sum()
controls_total_sa4["SA4"] = controls_total_sa4.index
```

```python
household_seed.to_csv('data/seed_households.csv')
person_seed.to_csv('data/seed_persons.csv')
controls_total_sa4.to_csv('data/controls_total_SA4.csv')
geo_cross_walk.to_csv('data/geo_cross_walk.csv')
```
